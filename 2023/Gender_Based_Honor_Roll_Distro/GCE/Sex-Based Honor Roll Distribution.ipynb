```python
import pandas as pd
import matplotlib.pyplot as plt

excel_file = "O_Level_Honor_Rolls.xlsx"  
df = pd.read_excel(excel_file, header=None, names=['SN', 'ExamCentre', 'CIN', 'CandidateName', 'Sex', 'Reg', 'Sat', 'Pass', 'Points'])

#S/N	ExamCentreName	CIN	Candidate Name	Sex	Reg	Sat	Pass	Points

gender_counts = df['Sex'].value_counts()

plt.figure(figsize=(8, 5))  
gender_counts.plot(kind='bar', color=['blue', 'yellow'], alpha=1.0)

plt.title('Male/Female GCE O-Level National Honor Roll Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend()

for i, count in enumerate(gender_counts):
    plt.text(i, count + 0.1, str(count), ha='left')

plt.show()
```
