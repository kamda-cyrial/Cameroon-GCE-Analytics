```python
import pandas as pd
import matplotlib.pyplot as plt

excel_file = "TVEE_Intermediate_Level_HR.xlsx"  
df = pd.read_excel(excel_file, header=None, names=['SN', 'ExamCentre', 'CIN', 'CandidateName', 'Sex', 'Speciality','Reg', 'Sat', 'Pass', 'Points'])

gender_counts = df['Sex'].value_counts()

plt.figure(figsize=(8, 5))  
gender_counts.plot(kind='bar', color=['blue', 'yellow'], alpha=1.0)

plt.title('Male/Female TVEE Intermediate Level National Honor Roll Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend()

for i, count in enumerate(gender_counts):
    plt.text(i, count + 0.1, str(count), ha='left')

plt.show()
```
