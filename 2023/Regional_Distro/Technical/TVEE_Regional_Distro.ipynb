```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data_file = "TVEE_Intermediate_Regional.xlsx"
dataframe = pd.read_excel(data_file)

plt.bar(dataframe['Region '], dataframe['% Pass'])

plt.title('Regional Distribution of TVEE Intermediate Level')
plt.xlabel('Region Name')
plt.ylabel('% Pass')

plt.show()
```

```python
data_file = "TVEE_Advanced_Level_Regional.xlsx"
dataframe = pd.read_excel(data_file)

plt.bar(dataframe['Region'], dataframe['% Pass'])

plt.title('Regional Distribution of TVEE Advanced Level')
plt.xlabel('Region')
plt.ylabel('% Pass')

plt.show()
```

```python
excel_file = "TVEE_Regional_Mix.xlsx"  
df = pd.read_excel(excel_file)

bar_width = 0.4 
index = np.arange(len(df['Region']))

plt.figure(figsize=(10, 5.5)) 
plt.bar(index - bar_width/2, df['%Pass_Intermediate_Level'], label='%Pass_Intermediate_Level', width=bar_width, color='yellow', alpha=1.0)
plt.bar(index + bar_width/2, df['%Pass_Advanced_Level'], label='%Pass_Advanced_Level', width=bar_width, color='blue', alpha=1.0)

plt.title('Regional Comparison of Success Rates in TVEE Intermediate and Advanced Levels')
plt.xlabel('Region')
plt.ylabel('Percentage')
plt.xticks(index, df['Region'])  
plt.legend()

plt.show()
```
