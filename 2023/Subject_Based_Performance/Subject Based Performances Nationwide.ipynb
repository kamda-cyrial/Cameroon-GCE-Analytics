```python
import pandas as pd
import matplotlib.pyplot as plt

excel_file = "Subject_Based_Performance_O_Level.xlsx"  
df = pd.read_excel(excel_file)

plt.figure(figsize=(15, 8))  
plt.bar(df['SubjectName'], df['% Pass'], color='blue', alpha=1.0)

plt.title('Percentage Pass by Subject GCE O Level')
plt.xlabel('Subject Name')
plt.ylabel('% Pass')
plt.xticks(rotation=45, ha='right')  

for i, percentage in enumerate(df['% Pass']):
    plt.text(i, percentage + 0.5, f"{percentage:.2f}%", ha='center')

plt.tight_layout() 
plt.show()
```
