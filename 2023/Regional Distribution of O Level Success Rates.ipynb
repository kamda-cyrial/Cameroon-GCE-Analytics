```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data_file = "O_Level_By_region_2023.xlsx"
dataframe = pd.read_excel(data_file)

plt.bar(dataframe['Region Name'], dataframe['% Pass'])

plt.title('Regional Distribution of GCE O Level Success Rates')
plt.xlabel('Region Name')
plt.ylabel('% Pass')

plt.show()
```

```python
data_file = "A_Level_Regional_Distro_2023.xlsx" 
dataframe = pd.read_excel(data_file)

plt.bar(dataframe['Region Name'], dataframe['% Pass'])

plt.title('Regional Distribution of GCE O Level Success Rates')
plt.xlabel('Region Name')
plt.ylabel('% Pass')

plt.show()
```

```python
excel_file = "Regional_Mix_2023.xlsx"  
df = pd.read_excel(excel_file)

bar_width = 0.4 
index = np.arange(len(df['Region']))

plt.figure(figsize=(10, 5.5)) 
plt.bar(index - bar_width/2, df['%pass_O_Level'], label='%pass_O_Level', width=bar_width, color='yellow', alpha=0.7)
plt.bar(index + bar_width/2, df['%pass_A_Level'], label='%pass_A_Level', width=bar_width, color='blue', alpha=0.7)

plt.title('Comparison of %pass_O_Level and %pass_A_Level by Region')
plt.xlabel('Region')
plt.ylabel('Percentage')
plt.xticks(index, df['Region'])  
plt.legend()

plt.show()
```
